```python
## Import Libraries
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()  ## Initiate h2o server
```

```python
## Reading The Data
data=h2o.import_file('/content/drive/MyDrive/Super_store.csv')
```

```python
data.head(5)
```

```python
## Drop Unrelated Features
data=data.drop(['Country','Row ID','Order ID'],axis=1)
```

```python
##Specifying The target & features
target='Profit'
features=data.columns.remove(target)
```

```python
## Spliting The dataset
train,test=data.split_frame(ratios=[0.8],seed=42)
```

```python
## Building & train automl

aml=H2OAutoML(max_models=20,seed=42,max_runtime_secs=360)   ## Training only 20 models or during 6 min total trainig time
aml.train(x=features,y=target,training_frame=train)
```

```python
## Dispalying Ranked models list with their metrics
print(aml.leaderboard)
```

```python
## Saving best model
best_model=aml.leader
```

```python
## Predicting The Test Data
preds=best_model.predict(test)
```

```python
## Showing best model metrics on train data
best_model.model_performance()
```

```python
## Showing best model metrics on test data
best_model.model_performance(test)
```

```python
r2_test=best_model.model_performance(test).r2()
r2_test
```
